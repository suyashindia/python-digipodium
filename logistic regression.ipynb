```python

#train a logistic regression classifier to predict whether flower is iris verginica or not 
from  sklearn import datasets
from sklearn.linear_model import LogisticRegression
import numpy as np
iris=datasets.load_iris()
print(list(iris.keys()))

 
```

```python
print(iris["data"])
print(iris["target"])
print(iris["DESCR"])
```

```python
print(iris["data"].shape)
```

```python
X=(iris["data"][:,3:])
y=(iris["target"]==2).astype(np.int)
print(y)
clf=LogisticRegression()
clf.fit(X,y)
example=clf.predict(([[2.6]]))
print(example) 

```
