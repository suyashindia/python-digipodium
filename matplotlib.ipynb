```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
```

practise,Line plot

```python
x=[0,1,2,3,4]
y=[0,2,5,6,8]
x2=np.arange(0,4,0.5)
plt.figure(figsize=(5,3),dpi=100)
print(x2)
plt.plot(x2,x2**2,color="green",linewidth=4)

plt.plot(x,y,label="2x",color="yellow",linewidth=4,marker=".",markersize=10,markeredgecolor="red",linestyle="solid")

plt.title("graph",fontdict={"fontname":"Comic Sans MS","fontsize": 20})#puts title
plt.xlabel("x axis",fontdict={"fontname":"Comic Sans MS","fontsize":20})
plt.ylabel("y label",fontdict={"fontname":"Comic Sans MS","fontsize":20})
plt.xticks([0,1,2,3])
plt.yticks([0,2,4,6,8,10])

plt.show()
```

bar plot

```python
labels=["A","B","C"]
values=[1,4,5]
color="green"
plt.bar(labels,values,color="green")
bars=plt.bar(labels,values)
bars[0].set_hatch("/")
bars[1].set_hatch("o")


plt.title("bargraph",fontdict={"fontname":"Comic Sans MS","fontsize": 20})
plt.xlabel("x axis",fontdict={"fontname":"Comic Sans MS","fontsize":20})
plt.ylabel("y label",fontdict={"fontname":"Comic Sans MS","fontsize":20})
```

advanced plotting

```python
data=pd.read_csv("https://raw.githubusercontent.com/digipodium/Datasets/main/house_pricing.csv")
```

```python
sns.lineplot(x="SquareFeet",y="Price",data=data)
plt.show()
```

```python
plt.plot(data.SquareFeet,data.Price)
plt.title("bargraph",fontdict={"fontname":"Comic Sans MS","fontsize": 20})
plt.show()
```

```python
for Squarefeet in data:
    if Squarefeet!="Price"
    plt.plot(data.Price,data["Squarefeet"])
```

```python
#histogram

plt.hist(data.Price,color="#00ffff")
plt.show()
```
